```python

df = pd.read_csv('./movies_metadata.csv')

def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return None

```

```python
df['production_companies'] = df['production_companies'].apply(lambda x: safe_literal_eval(x) if pd.notnull(x) else None)
df['num_companies'] = df['production_companies'].apply(lambda x: len(x) if x is not None else 0)
max_companies = max(df['num_companies'])

for i in range(1, max_companies + 1):
    col_name = f'production_companies_{i}'
    df[col_name] = df['production_companies'].apply(lambda x: x[i-1]['name'] if x is not None and i <= len(x) else '')
df.head()
```

```python
df['genre'] = df['genres'].apply(lambda x: safe_literal_eval(x) if isinstance(x, str) and x.strip() != '' else None)
df['num_genre'] = df['genre'].apply(lambda x: len(x) if x is not None else 0)
max_genre = max(df['num_genre'])

for i in range(1, max_genre + 1):
    col_name = f'genre_{i}'
    df[col_name] = df['genre'].apply(lambda x: x[i-1]['name'] if isinstance(x, list) and len(x) >= i else '')

# df = df.drop(['genres', 'genre', 'num_genre'], axis=1)
```

```python
df['genre'] = df['genres'].apply(lambda x: safe_literal_eval(x) if isinstance(x, str) and x.strip() != '' else None)
df['genre'] = df['genre'].map(lambda x: [d['name'] for d in x] if x is not None else None)
```

```python
df['production_companies_name'] = df['production_companies'].apply(lambda x: safe_literal_eval(x) if isinstance(x, str) and x.strip() != '' else None)
df['production_companies_name'] = df['production_companies'].map(lambda x: [d['name'] for d in x] if x is not None else None)
```

```python
df.to_csv('movmat_entities.csv', index=False)
df.head()
```
